```python
# Import necessary libraries 
import pandas as pd
```

```python
# Load the dataset from the CSV file
df = pd.read_csv('../data/raw/Activities.csv')
```

```python
# Display the first few rows of the DataFrame to understand its structure
df.head()
```

```python
# Get a concise summary of the DataFrame, including the number of non-null values in each column
df.info()
```

```python
# Convert Date & Time to correct format
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d %H:%M:%S')
df.info()
```

```python
# Display statistical summaries of numeric columns
df.describe()
```

```python
def generate_alphanumeric_id(prefix, idx):
    """
    Generates an alphanumeric ID with a given prefix and index.
    
    Parameters:
    - prefix (str): The prefix for the ID.
    - idx (int): The numeric index for the ID.
    
    Returns:
    - str: An alphanumeric ID.
    """
    return f"{prefix}{idx:03d}"  # Pads the index with zeros, e.g., AT001

# Generate unique IDs for each activity type
activity_types = df['Activity Type'].unique()

# Generate Alphanumeric IDs
alphanumeric_ids = [generate_alphanumeric_id("AT", i+1) for i in range(len(activity_types))]

# Create a mapping DataFrame
activity_types_df = pd.DataFrame({
    'ActivityTypeID': alphanumeric_ids,
    'ActivityType': activity_types
})


# Display the Activity Types table
activity_types_df
```

```python
# Export Activty Types to csv
activity_types_df.to_csv('../data/processed/activity_types.csv', index=False)
```

```python
# Drop columns where all values are 0
df = df.loc[:, (df != 0).any(axis=0)]

df.info()
```

```python
# Add an Activity ID column as a primary key
df['ActivityID'] = range(1, len(df) + 1)
df.tail()
```

```python
# Merge df with activity_types_df on "Activity Type" to include "ActivityTypeID"
merged_df = pd.merge(df, activity_types_df, 
                         left_on='Activity Type', right_on='ActivityType', 
                         how='left')

# Drop the redundant 'ActivityType' column from the merged DataFrame
merged_df.drop('ActivityType', axis=1, inplace=True)

# Now, activities_df is updated to contain ActivityID, Activity Type, Date, Title, and ActivityTypeID
merged_df.head(10)
```

```python
# Create the Activities DataFrame
activities_df = merged_df[['ActivityID','ActivityTypeID', 'Date', 'Title']].copy()
activities_df.head(10)
```

```python
# Export Activty Types to csv
activities_df.to_csv('../data/processed/activities.csv', index=False)
```
