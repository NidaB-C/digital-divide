```python
import pandas as pd
import sqlite3
import os
import pandas as pd
```

```python
# Load csv file into pandas dataframe
df = pd.read_csv('../data/processed/performance_metrics.csv')
df
```

```python
# Creating SQLite database
conn = sqlite3.connect('database.db')
cursor = conn.cursor()
```

```python
# Table Definition
create_table = '''CREATE TABLE IF NOT EXISTS ActivityTypes(
                ActivityTypeID Text PRIMARY KEY,
                ActivityType Text NOT NULL);
                '''

# Creating the table into our database
cursor.execute(create_table)
```

```python
# insert the data from the DataFrame into the SQLite table
df.to_sql('ActivityTypes', conn, if_exists='replace', index = False)

# Printing pandas dataframe
pd.read_sql('''SELECT * FROM ActivityTypes''', conn)
```

```python
x_axis = df["Distance"]
y_axis = df["Calories"]

from bokeh.plotting import figure, output_file, save, show
from bokeh.resources import CDN
from bokeh.embed import file_html


# set output to static HTML file
output_file(filename="../data/bokeh/test_output.html", title="Bokeh Test Output")

p = figure(width=400, height=400)

# add a circle renderer with a size, color, and alpha
p.circle(x_axis, y_axis, size=20, color="navy", alpha=0.5)

# show the results
show(p)

# save the results to a file
save(p)
```
