```python
import sqlite3
import pandas as pd
```

```python
# Path to your SQLite database
db_path = '../database/activity_tracking.db'
```

```python
# Establish a connection to the SQLite database
conn = sqlite3.connect(db_path)
```

```python
# Function to display a table
def display_table(conn, table_name):
    query = f"SELECT * FROM {table_name}"
    df = pd.read_sql_query(query, conn)
    display(df)
```

```python
#Display the 'Activities' table
display_table(conn, 'Activities')
```

```python
#Display the 'ActivityTypes' table
display_table(conn, 'ActivityTypes')
```

```python
# Close the connection 
conn.close()
```
